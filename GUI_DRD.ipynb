```python
## For GUI
from tkinter import *
import tkinter
import os
from tkinter.font import Font
from PIL import Image, ImageTk as itk
from tkinter import ttk
from tkinter import filedialog as fd
from tkinter.messagebox import showinfo

## For Saved Model
import numpy as np
from keras.models import load_model
```

```python
def arrow_multi():
    
    saved_model_multi = load_model("E:/Kushagra/College/SEM 7/Project/Final Models/multiclass.h5")
    
    # Assigning label names to the corresponding indexes
    labels = {0: 'Mild DR', 1: 'Moderate DR', 2: 'No DR', 3:'Proliferate DR', 4: 'Severe DR'}

    image = []
    img= Image.open(filename1)
    # resizing the image to (256,256)
    img = img.resize((256,256))
    # appending image to the image list
    image.append(img)
    # converting image to array
    img = np.asarray(img, dtype= np.float32)
    # normalizing the image
    img = img / 255
    # reshaping the image in to a 4D array
    img = img.reshape(-1,256,256,3)
    # making prediction of the model
    predict = saved_model_multi.predict(img)
    # getting the index corresponding to the highest value in the prediction
    predict = np.argmax(predict)
    # appending the predicted class to the list
    prediction = labels[predict]
    
    if ("No" in filename1):
        original = "No DR"
    elif ("Mild" in filename1):
        original = "Mild DR"
    elif ("Moderate" in filename1):
        original = "Moderate DR"
    elif ("Severe" in filename1):
        original = "Severe DR"
    elif ("Proliferate" in filename1):
        original = "Proliferate DR"
    else:
        original = "Wrong Input"
        
    label2["state"] = NORMAL
    label3["state"] = NORMAL
        
    label4 = Label(screen1, text = prediction.upper(), fg = "black", width = 15, bg = "white", font=('Calibri',28,"bold"))
    label4.place(x=1400,y=550)
    label5 = Label(screen1, text = original.upper(), fg = "black", width = 15, bg = "white", font=('Calibri',28,"bold"))
    label5.place(x=1400,y=850)
```

```python
def choose_image_multi():
    global filename1
    filetypes = [('Png Files', '*.png')]
    filename1 = fd.askopenfilename(title='Select an Image', initialdir='E:/Kushagra/College/SEM 7/Project/Dataset/', filetypes=filetypes)
    showinfo(title='Selected Image...', message="Opening : "+filename1)
    img = Image.open(filename1)
    img_resized=img.resize((400,400))
    img=itk.PhotoImage(img_resized)
    button3 = Button(image = img, width = "400", height = "400", bd = 2, bg = "black")
    button3.image = img
    button3.place(x=235,y=450)
    state1()
```
